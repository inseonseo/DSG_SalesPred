__종속변수(y) 스케일링__ (by ChatGPT) 
- 대부분의 회귀 문제에서는 종속변수를 스케일링하지 않아도 됨  
- 상황에 따라 종속변수 스케일링이 도움이 될 수 있음.  
    - __다변량 종속변수__: 종속변수가 다변량 데이터(다중 타겟)일 때는 각 종속변수 간의 스케일이 크게 다르면 모델이 특정 종속변수에 지나치게 영향을 받을 수 있음
    - __극단적인 값이나 이상치가 포함된 경우__: 종속변수에 극단적인 값이나 이상치가 포함된 경우, 스케일링을 통해 모델의 안정성과 성능을 향상시킬 수 있습니다.

➡️ 종속변수(train) 데이터는 이상치가 포함되어 있으므로, MinMaxScaling을 진행한다.
